### Bu yÄ±la ait verileri alÄ±n

Her iki yÄ±lÄ± karÅŸÄ±laÅŸtÄ±rabilmek iÃ§in bu yÄ±la ait verileri de eklemeniz gerekir.
Bu verileri zaten 4. bÃ¶lÃ¼mde hazÄ±rladÄ±nÄ±z.

ðŸ“Œ BÃ¶lÃ¼m 4'teki "names", "call_numbers", "average_deal_sizes" ve "revenues" NumPy dizilerini kopyalayÄ±p yapÄ±ÅŸtÄ±rÄ±n.

```python
#Copy and paste the NumPy arrays "names", "call_numbers", "average_deal_sizes", and "revenues"
names = ['Ben', 'Omer', 'Karen', 'Celine', 'Sue', 'Bora', 'Rose', 'Ellen', 'Bob', 'Taylor,', 'Jude']
call_numbers = [300, 10, 500, 70, 100, 100, 600, 800, 200, 450, 80]
average_deal_sizes=[8, 6, 24, 32, 5, 25, 25, 40, 15, 10, 12]
revenues = [2400, 60, 12000, 2275, 500, 770, 4000, 6000, 800, 1200, 500]
```

Åžimdi bu dizileri kullanarak bir DataFrame oluÅŸturun.

ðŸ“Œ Dizilerle bir sÃ¶zlÃ¼k oluÅŸturun. Anahtarlardaki sÃ¼tun adlarÄ±nÄ± belirtin.

ðŸ“Œ SÃ¶zlÃ¼ÄŸÃ¼ bir Pandas DataFrame "current_year" biÃ§imine dÃ¶nÃ¼ÅŸtÃ¼rÃ¼n ve kontrol etmek iÃ§in .head() iÅŸlevini kullanÄ±n.

```python
#Create a dictionary with the column names as keys and the arrays as values
dictionary = {"name" :names,
       "calls":call_numbers,
       "avg_deal_size":average_deal_sizes,
       "revenue":revenues}
#Convert the dictionary to a Pandas Dataframe
current_year = pd.DataFrame(dictionary)
#Use the .head() function to check it
current_year.head()
```

GeÃ§en yÄ±lÄ±n verileriyle yaptÄ±ÄŸÄ±mÄ±za benzer ÅŸekilde yÄ±l bilgilerini ekleyin.

ðŸ“Œ "YÄ±l" sÃ¼tununu DataFrame'e ekleyin ve "2022" deÄŸerini satÄ±rlarÄ±na aktarÄ±n. Daha sonra sonucu gÃ¶rmek iÃ§in .head() iÅŸlevini kullanÄ±n.

```python
#Add the column "Year" and assign the value "2022" to its rows
current_year["Year"] = 2022
#Use the .head() function to see the result
current_year.head()
```

###Ä°ki DataFrame'i karÅŸÄ±laÅŸtÄ±rÄ±n

ArtÄ±k DataFrame "current_year"Ä± yazdÄ±rdÄ±ÄŸÄ±nÄ±za gÃ¶re, bunlarÄ± karÅŸÄ±laÅŸtÄ±rmak iÃ§in "last_year"Ä± da yazdÄ±rÄ±n.

### Ä°ki DataFrame'i birleÅŸtirin
ArtÄ±k iki DataFrame aynÄ± sÃ¼tun adlarÄ±na sahip olduÄŸundan, bunlarÄ± tek bir DataFrame "all_data" iÃ§inde birleÅŸtirebilir veya birleÅŸtirebilirsiniz.

```python
#Use the .concat() function to concatenate the two DataFrames
all_data = pd.concat([last_year,current_year],axis = 0)
#Display the DataFrame "all_data"
all_data
```

###Verileri kontrol edin

Bu iÅŸe yaradÄ± ancak dizinlerin yanlÄ±ÅŸ olduÄŸunu fark ettiniz. BunlarÄ± sÄ±fÄ±rlamanÄ±z gerekir.

ðŸ“Œ .reset_index() fonksiyonunu kullanÄ±n. "DÃ¼ÅŸme" ve "yerinde" baÄŸÄ±msÄ±z deÄŸiÅŸkenlerini "DoÄŸru" olarak ayarlayÄ±n. ArdÄ±ndan DataFrame'i gÃ¶rÃ¼ntÃ¼leyin.

```python
#Use the .reset_index() function to reset the indexes
all_data.reset_index(drop =True, inplace = True)
#Display the DataFrame "all_data"
all_data
```

Next, you need to check the entries for missing values.

ðŸ“Œ Use the isna.() and .any() function to see if there are missing values in the DataFrame.

```python
#Check the DataFrame for missing values
all_data.isna().any()
```

Ã‡Ä±ktÄ±da "Ortalama anlaÅŸma boyutu" ve "Gelir" sÃ¼tunlarÄ±nda eksik deÄŸerler olduÄŸunu gÃ¶rÃ¼yorsunuz. Eksik deÄŸerleri ilgili sÃ¼tunun ortalamasÄ±yla doldurarak bu sorunu Ã§Ã¶zmeye karar veriyorsunuz.

ðŸ“Œ Eksik deÄŸerleri doldurmak iÃ§in .fillna() fonksiyonunu kullanÄ±n. "DeÄŸer" baÄŸÄ±msÄ±z deÄŸiÅŸkenini "tÃ¼m_verilerin" ortalamasÄ±na ayarlamak iÃ§in .mean() iÅŸlevini kullanÄ±n. Yine "yerinde" seÃ§eneÄŸini "DoÄŸru" olarak ayarlayÄ±n.
