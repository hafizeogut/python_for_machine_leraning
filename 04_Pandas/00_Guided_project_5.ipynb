### Bu yıla ait verileri alın

Her iki yılı karşılaştırabilmek için bu yıla ait verileri de eklemeniz gerekir.
Bu verileri zaten 4. bölümde hazırladınız.

📌 Bölüm 4'teki "names", "call_numbers", "average_deal_sizes" ve "revenues" NumPy dizilerini kopyalayıp yapıştırın.

```python
#Copy and paste the NumPy arrays "names", "call_numbers", "average_deal_sizes", and "revenues"
names = ['Ben', 'Omer', 'Karen', 'Celine', 'Sue', 'Bora', 'Rose', 'Ellen', 'Bob', 'Taylor,', 'Jude']
call_numbers = [300, 10, 500, 70, 100, 100, 600, 800, 200, 450, 80]
average_deal_sizes=[8, 6, 24, 32, 5, 25, 25, 40, 15, 10, 12]
revenues = [2400, 60, 12000, 2275, 500, 770, 4000, 6000, 800, 1200, 500]
```

Şimdi bu dizileri kullanarak bir DataFrame oluşturun.

📌 Dizilerle bir sözlük oluşturun. Anahtarlardaki sütun adlarını belirtin.

📌 Sözlüğü bir Pandas DataFrame "current_year" biçimine dönüştürün ve kontrol etmek için .head() işlevini kullanın.

```python
#Create a dictionary with the column names as keys and the arrays as values
dictionary = {"name" :names,
       "calls":call_numbers,
       "avg_deal_size":average_deal_sizes,
       "revenue":revenues}
#Convert the dictionary to a Pandas Dataframe
current_year = pd.DataFrame(dictionary)
#Use the .head() function to check it
current_year.head()
```

Geçen yılın verileriyle yaptığımıza benzer şekilde yıl bilgilerini ekleyin.

📌 "Yıl" sütununu DataFrame'e ekleyin ve "2022" değerini satırlarına aktarın. Daha sonra sonucu görmek için .head() işlevini kullanın.

```python
#Add the column "Year" and assign the value "2022" to its rows
current_year["Year"] = 2022
#Use the .head() function to see the result
current_year.head()
```

###İki DataFrame'i karşılaştırın

Artık DataFrame "current_year"ı yazdırdığınıza göre, bunları karşılaştırmak için "last_year"ı da yazdırın.

📌 "last_year" DataFrame'i yazdırmak için .head() işlevini kullanın.

```python
#Use the .head() function to print the DataFrame "last_year"
last_year.head()
```

You notice that the column names of the two DataFrames are different. You need to fix this problem.

📌 Assign the column names of "last_year" to "current_year" by using the *columns* attribute.

```python
#Use the columns attribute to assign the column names of "last_year" to "current_year"
current_year.columns = last_year.columns
```

### İki DataFrame'i birleştirin
Artık iki DataFrame aynı sütun adlarına sahip olduğundan, bunları tek bir DataFrame "all_data" içinde birleştirebilir veya birleştirebilirsiniz.

📌 .concat() fonksiyonunu "axis" argümanını 0 olarak ayarlayarak kullanın. Ardından DataFrame'i görüntüleyin.

```python
#Use the .concat() function to concatenate the two DataFrames
all_data = pd.concat([last_year,current_year],axis = 0)
#Display the DataFrame "all_data"
all_data
```

###Verileri kontrol edin

Bu işe yaradı ancak dizinlerin yanlış olduğunu fark ettiniz. Bunları sıfırlamanız gerekir.

📌 .reset_index() fonksiyonunu kullanın. "Düşme" ve "yerinde" bağımsız değişkenlerini "Doğru" olarak ayarlayın. Ardından DataFrame'i görüntüleyin.

```python
#Use the .reset_index() function to reset the indexes
all_data.reset_index(drop =True, inplace = True)
#Display the DataFrame "all_data"
all_data
```

Next, you need to check the entries for missing values.

📌 Use the isna.() and .any() function to see if there are missing values in the DataFrame.

```python
#Check the DataFrame for missing values
all_data.isna().any()
```

Çıktıda "Ortalama anlaşma boyutu" ve "Gelir" sütunlarında eksik değerler olduğunu görüyorsunuz. Eksik değerleri ilgili sütunun ortalamasıyla doldurarak bu sorunu çözmeye karar veriyorsunuz.

📌 Eksik değerleri doldurmak için .fillna() fonksiyonunu kullanın. "Değer" bağımsız değişkenini "tüm_verilerin" ortalamasına ayarlamak için .mean() işlevini kullanın. Yine "yerinde" seçeneğini "Doğru" olarak ayarlayın.

```python
#Replace the missing values with the mean of the respective column
all_data.fillna(value = np.mean(all_data),inplace = True)
#Display the DataFrame
all_data
```

Ayrıca bazı kopya satırlar da olabilir, bunları kaldırmanız gerekir.

📌 Yinelenen satırları kaldırmak için .drop_duplicates() yöntemini kullanın.

📌reset_index() fonksiyonunu kullanarak indeksleri tekrar sıfırlayın.

```python
#Drop the duplicates
all_data.drop_duplicates(inplace=True) 
#Display the DataFrame 
all_data
```

```python
all_data.reset_index(drop = True )
all_data
```
