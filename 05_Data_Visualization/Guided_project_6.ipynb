###Matplotlib Temelleri

GÃ¶revinize odaklanmadan Ã¶nce Matplotlib'in temellerini uygulayÄ±n.

ðŸ“Œ Ã–rnek veriler oluÅŸturun:
   1. X ekseni deÄŸerlerinin listesi
   2. Y ekseni deÄŸerlerinin listesi

```python
#Create a list for the x-axis values
x = [0,2,4,6,8,10,12,14,16]
#Create a list for the y-axis values
y = [ 0,4,16,36,64,100,144,196,256]
```

####Ã‡izgi grafiÄŸi
FarklÄ± grafik tÃ¼rlerinden bazÄ±larÄ±nÄ± keÅŸfetmek iÃ§in bu basit verileri kullanÄ±n.

ðŸ“Œ Ã‡izgi grafiÄŸi oluÅŸturmak iÃ§in Matplotlib'in .plot() fonksiyonunu, "Ã–rnek veri - Ã‡izgi grafiÄŸi" baÅŸlÄ±ÄŸÄ±nÄ± eklemek iÃ§in .title() metodunu kullanÄ±n.

ðŸ“Œ GrafiÄŸi gÃ¶rÃ¼ntÃ¼lemek iÃ§in *plt.show()* kullanÄ±n.

```python
#Use the .plot() function to create a line plot
plt.plot(x,y)
#Use the .title() method to add the title
plt.title("Example data - Line plot")
#Display the plot
plt.show()
```

####DaÄŸÄ±lÄ±m grafiÄŸi

Daha sonra, bir daÄŸÄ±lÄ±m grafiÄŸi oluÅŸturmak iÃ§in aynÄ± verileri kullanÄ±n.

ðŸ“Œ Bir daÄŸÄ±lÄ±m grafiÄŸi oluÅŸturmak iÃ§in Matplotlib'in .scatter() fonksiyonunu kullanÄ±n ve "Ã–rnek veri - DaÄŸÄ±lÄ±m grafiÄŸi" baÅŸlÄ±ÄŸÄ±nÄ± eklemek iÃ§in .title() yÃ¶ntemini kullanÄ±n.

ðŸ“Œ GrafiÄŸi gÃ¶rÃ¼ntÃ¼lemek iÃ§in *plt.show()* kullanÄ±n.

```python
#Use the .scatter() function to create a scatter plot
plt.scatter(x,y)
#Use the .title() method to add the title
plt.title("Example data - Line plot")

#Display the plot
plt.show()
```

####Grafik Ã§ubuÄŸu

Son olarak, bir Ã§ubuk grafik oluÅŸturmak iÃ§in aynÄ± verileri kullanÄ±n.

ðŸ“Œ Ã‡ubuk grafik oluÅŸturmak iÃ§in Matplotlib'in .bar() fonksiyonunu kullanÄ±n ve "Ã–rnek veri - Ã‡ubuk grafik" baÅŸlÄ±ÄŸÄ±nÄ± eklemek iÃ§in .title() yÃ¶ntemini kullanÄ±n.

ðŸ“Œ GrafiÄŸi gÃ¶rÃ¼ntÃ¼lemek iÃ§in *plt.show()* kullanÄ±n.

```python
#Use the .bar() function to create a bar chart
plt.bar(x,y)

#Use the .title() method to add the title
plt.title("Example data - Line plot")

#Display the chart
plt.show()
```

Grafikleri yan yana gÃ¶rmek istediÄŸinize karar veriyorsunuz. Åžeklin daha okunaklÄ± olmasÄ± iÃ§in her grafiÄŸin farklÄ± bir rengi olmalÄ±dÄ±r.

ðŸ“Œ 18x5 boyutunda bir ÅŸekil nesnesi oluÅŸturmak iÃ§in Matplotlib'in .figure() fonksiyonunu ve onun "figsize" argÃ¼manÄ±nÄ± kullanÄ±n.

ðŸ“Œ Yeni oluÅŸturduÄŸunuz Ã¼Ã§ grafiÄŸi ÅŸekle eklemek ve her birine bir baÅŸlÄ±k eklemek iÃ§in .add_subplot() yÃ¶ntemini kullanÄ±n.

   * 3 grafiÄŸin bulunduÄŸu 1 satÄ±r olmalÄ±dÄ±r.

   * Her grafiÄŸin rengini deÄŸiÅŸtirmek iÃ§in "renk" argÃ¼manÄ±nÄ± kullanÄ±n:
     1. Ã‡izgi grafiÄŸi: kÄ±rmÄ±zÄ±
     2. DaÄŸÄ±lÄ±m grafiÄŸi: yeÅŸil
     3. Ã‡ubuk grafik: turuncu

```python
#Create a figure object of the size 18x5
fig =plt.figure(figsize=(18,5))

#Use the .add_subplot() method to add the line plot
first_plot = fig.add_subplot(1,3,1)

#Change the color to red
first_plot.plot(x,y,color = "red")

#Add the title
first_plot.set_title("Example data - Line plot")

#Use the .add_subplot() method to add the scatter plot
second_plot = fig.add_subplot(1,3,2)

#Change the color to green
second_plot.scatter(x,y,color="green")

#Add the title
second_plot.set_title("Example data - Line plot")

#Use the .add_subplot() method to add the bar chart
third_plot = fig.add_subplot(1,3,3)

#Change the color to orange
third_plot.bar(x,y,color = "orange")
#Add the title
third_plot.set_title("Example data - Line plot")
#Display the figure
plt.show
```

###SatÄ±ÅŸ raporunun gÃ¶rselleÅŸtirilmesi

ArtÄ±k gÃ¶revinizi Ã¼stlenmeye hazÄ±rsÄ±nÄ±z.

SatÄ±ÅŸ ekibi bir rapor oluÅŸturuyor ve sonuÃ§larÄ±n gÃ¶rselleÅŸtirilmesine ihtiyaÃ§ duyuyor. YÃ¶neticiniz Rachel, raporu inceleyen kiÅŸilerin verileri daha iyi anlayabilmesi iÃ§in sizden bazÄ± grafikler oluÅŸturmanÄ±zÄ± istiyor. Size satÄ±ÅŸ verilerini iÃ§eren bir .csv dosyasÄ± sunarlar.

####Veri HazÄ±rlama

Ã–ncelikle verileri .csv dosyasÄ±ndan almanÄ±z gerekir.

ðŸ“Œ "employee_performans.csv" dosyasÄ±nÄ± okumak ve bunu "data" deÄŸiÅŸkenine atamak iÃ§in .read_csv() fonksiyonunu kullanÄ±n.

ðŸ“Œ Veri kÃ¼mesinin ne iÃ§erdiÄŸini kontrol etmek iÃ§in .head() iÅŸlevini kullanÄ±n.

```python
#Read the data from "employee_performance.csv""
data = pd.read_csv("employee_performance.csv")

#Use the .head() function to check the dataset
data.head()
```

####EÄŸitim dÃ¼zeyinin grafiÄŸini Ã§izin

Ã–ncelikle satÄ±ÅŸ ekibinin eÄŸitim seviyesi hakkÄ±nda bazÄ± Ã¶ngÃ¶rÃ¼ler oluÅŸturuyorsunuz ve pasta grafiÄŸinin faydalÄ± olacaÄŸÄ±na karar veriyorsunuz. FarklÄ± kategorilerdeki kiÅŸi sayÄ±sÄ±nÄ± belirlemeniz gerekiyor.

ðŸ“Œ FarklÄ± kategorilerdeki kiÅŸi sayÄ±sÄ±nÄ± belirlemek iÃ§in .value_counts() fonksiyonunu kullanÄ±n.

```python
#Use the .value_counts() function to determine the number of people in the different categories
educational_level = data["Education"].value_counts()

#Display the result
educational_level
```

ArdÄ±ndan "Ãœniversite", "Lise", "Ãœniversite" etiketlerini iÃ§eren bir pasta grafiÄŸi oluÅŸturun.

ðŸ“Œ Pasta grafiÄŸi oluÅŸturmak iÃ§in .pie() fonksiyonunu ve Pandas Serisinin indekslerini etiket olarak vermek iÃ§in "labels" argÃ¼manÄ±nÄ± kullanÄ±n.

```python
#Create a pie chart with the labels "College", "High School", "University"
plt.pie(educational_level,labels=educational_level.index)

#Display the chart
plt.show()
```

####Gelir grafiÄŸini Ã§izin

Daha sonra Ã§alÄ±ÅŸanlarÄ±n elde ettiÄŸi gelirin Ã§ubuk grafiÄŸini oluÅŸturmanÄ±z gerekir.
Adlar x ekseninde, gelirler ise y ekseninde olmalÄ±dÄ±r.

ðŸ“Œ Gelir deÄŸerlerinin Ã§ubuk grafiÄŸini oluÅŸturmak iÃ§in .bar() iÅŸlevini kullanÄ±n.

```python
#Create a bar chart with the names on the x-axis and the revenue values on the y-axis
plt.bar(data["Name"],data["Revenue"])

#Display the chart
plt.show()
```

Bu durumdaki grafik yalnÄ±zca Ã§alÄ±ÅŸanlar arasÄ±ndaki karÅŸÄ±laÅŸtÄ±rmayÄ± gÃ¶sterir.

AnlaÅŸÄ±labilirliÄŸi artÄ±rmak iÃ§in grafiÄŸe "Ã‡aÄŸrÄ± sayÄ±sÄ±" verilerini eklemeye karar veriyorsunuz.

ðŸ“Œ "Gelir" ve "Ã‡aÄŸrÄ± sayÄ±sÄ±" verilerini Ã§ubuk grafiÄŸe ekleyin. Verileri etiketlemek iÃ§in "etiket" baÄŸÄ±msÄ±z deÄŸiÅŸkenini kullanÄ±n.

Verileri ayÄ±rt etmek iÃ§in bir gÃ¶sterge eklemeniz gerekir. AyrÄ±ca Ä±zgara Ã§izgileri eklemek de faydalÄ± olacaktÄ±r.

ðŸ“Œ Izgara Ã§izgileri eklemek iÃ§in .legend() fonksiyonunu ve .grid() fonksiyonunu kullanarak bir efsane ekleyin.;

```python
#Create a bar chart with the data "Revenue" and "Number of calls"
plt.bar(data["Name"],data["Revenue"],label = "Revenues")
plt.bar(data["Name"],data["Number of Calls"],label = "Number of calls")

#Add a legend
plt.legend()

#Add grid lines
plt.grid()

#Display the chart
plt.show()
```
